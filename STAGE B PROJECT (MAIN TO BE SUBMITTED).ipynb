## AIM ##
The aim of this is to be able to predict the energy use by appliances in the dataset. Therefore, the attribute "Appliances" is the target. We'll achieve this by adopting linear regression model on the dataset.
We'll also try to regularize the model for better results.
Our metric tests are:
##### Mean Absolute Error (MAE) #####
##### R-Squared #####
##### Residual Sum of Squares #####
##### Root Mean Square Error #####


```python
import pandas as pd
#reading the dataset using pandas
df = pd.read_csv('energydata_complete.csv')
#description of the attributes/features of the dataset
df.describe()
```

```python
df
```

```python
df.info()
```

```python
#Dropping columns "lights" and "date" as stated in the instruction

df.drop(['date', 'lights'], axis = 1, inplace = True)
```

```python
df.info()
```

```python
# Normalising the Dataset and seperating the target column from the rest of the features
# Using the MinMaxScaler() for normalisation
# Naming the attributes without the target column "features_df"
# Naming the target column "target"

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
normalised_df = pd.DataFrame(scaler.fit_transform(df), columns = df.columns)
target = normalised_df['Appliances']
features_df = normalised_df.drop(columns = ['Appliances'])
```

```python
features_df.describe()
```

```python
target.describe()
```

```python
# Splitting of the data set into Train and Test data respectively using the train_test_split module
# Also setting the test_size to 30percent and thus the train_size to 70%
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features_df, target, test_size = 0.3, random_state = 42)
```

```python
# Linear Regression Model
# Here we make use of the multivariate regression model
# We fit our model on the Train data then predict it on the 
# test data that we splitted into in the previous cell

from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()
linear_model.fit(X_train, y_train)
predicted_values = linear_model.predict(X_test)
```

##### Applying our Metric Tests on model used #####
This is done in order to measure the regression performance.


```python
# Mean Absolute Error(MAE) Linear
# The smaller the MAE is the better the model, Our MAE should be close to 0

from sklearn.metrics import mean_absolute_error 
mae = mean_absolute_error(y_test, predicted_values)
round(mae, 2)
```

```python
# Residual Sum of Squares (RSS) Linear
# Explains how well the model approximates the data
# The lower the RSS, the better the model's estimations

import numpy as np
rss = np.sum(np.square(y_test - predicted_values))
round(rss, 2)
```

```python
# R_Squared linear
# ranges from 0 to 1
# the higher the value, the better the model.

from sklearn.metrics import r2_score
r2_score = r2_score(y_test, predicted_values)
round(r2_score, 2)
```

```python
# Root Mean Square Error Linear
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, predicted_values))
round(rmse, 3)
```

## REGULARISATION ##


### Lasso Regression ###

```python
#REGULARISATION

#Lasso Regression
from sklearn.linear_model import Lasso
lasso_reg = Lasso(alpha = 0.001)
lasso_reg.fit(X_train, y_train)
predicted_lasso = lasso_reg.predict(X_test)
```

```python
# R-Squared Lasso
from sklearn.metrics import r2_score
score = r2_score(y_test, predicted_lasso)
round(score, 3)
```

```python
# RMSE Lasso
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, predicted_lasso))
round(rmse, 3)
```

```python
# RSS Lasso
import numpy as np
rss = np.sum(np.square(y_test - predicted_lasso))
round(rss, 3)
```

```python
# MAE Lasso
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, predicted_lasso)
round(mae, 3)
```

### Ridge Regression ###

```python
from sklearn.linear_model import Ridge 
ridge_reg = Ridge(alpha = 0.4)
ridge_reg.fit(X_train, y_train)
predicted_ridge = ridge_reg.predict(X_test)
```

```python
# R-Squared Ridge
from sklearn.metrics import r2_score
score = r2_score(y_test, predicted_ridge)
round(score, 3)
```

```python
# RMSE Ridge
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, predicted_ridge))
round(rmse, 3)
```

```python
# RSS Ridge
import numpy as np
rss = np.sum(np.square(y_test - predicted_ridge))
round(rss, 3)
```

```python
# MAE Ridge
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, predicted_ridge)
round(mae, 3)
```

### REGULARISATION COMPARISON ###

```python
# Comparing the Effects of regularisation using the weights of the coefficients of various models
def get_weights_df(model, feat, col_name):
    weights = pd.Series(model.coef_, feat.columns).sort_values()
    weights_df = pd.DataFrame(weights).reset_index()
    weights_df.columns = ['Features', col_name]
    weights_df[col_name].round(3)
    return weights_df

linear_model_weights = get_weights_df(linear_model, X_train, 'Linear_Model_Weight')
ridge_weights = get_weights_df(ridge_reg, X_train, 'Ridge_Weight')
lasso_weights = get_weights_df(lasso_reg, X_train, 'Lasso_Weight')


final_weights = pd.merge(linear_model_weights, ridge_weights, on = 'Features')
final_weights = pd.merge(final_weights, lasso_weights, on = 'Features')
final_weights
```
